```python
from pyspark.sql import SparkSession

# 1. Initialize Spark Session
spark = SparkSession.builder \
    .appName("LinearRegression_Bus_Step1") \
    .getOrCreate()

# 2. Load the Dataset
# We use 'inferSchema=True' so Spark automatically recognizes numbers (Integer/Double) vs text (String)
df = spark.read.csv("medium_bus_data.csv", header=True, inferSchema=True)

# 3. Inspection
print(f"✅ Data Loaded Successfully!")
print(f"Total Rows: {df.count()}")

print("\n--- Data Schema (Column Types) ---")
df.printSchema()

print("\n--- Data Preview (First 5 Rows) ---")
df.show(5)
```

```python
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler, StandardScaler
from pyspark.ml import Pipeline
from pyspark.sql.functions import col

print("=== STEP 2: DATA PREPROCESSING & FEATURE ENGINEERING ===\n")

# 1. Define Columns to Transform
# We need to convert text columns (Categorical) into numbers
categorical_cols = ["Day_of_Week", "LineName", "Route_Type"]

# Output column names
index_cols = [c + "_Idx" for c in categorical_cols]
vec_cols = [c + "_Vec" for c in categorical_cols]

# 2. Build the Preprocessing Stages
stages = []

# A. StringIndexer: Convert Text -> Number Index (e.g., "Monday" -> 0)
# We loop through each column and create an indexer
for col_name, idx_name in zip(categorical_cols, index_cols):
    indexer = StringIndexer(inputCol=col_name, outputCol=idx_name)
    stages.append(indexer)

# B. OneHotEncoder: Convert Index -> Binary Vector (e.g., 0 -> [1, 0, 0])
# This is crucial for Linear Regression to handle categories correctly
encoder = OneHotEncoder(inputCols=index_cols, outputCols=vec_cols)
stages.append(encoder)

# C. VectorAssembler: Combine all features into one vector
# Inputs: The encoded vectors + Numerical columns (Hour, Traffic Index)
assembler_inputs = vec_cols + ["Hour_of_Day", "Live_Traffic_Index"]
assembler = VectorAssembler(inputCols=assembler_inputs, outputCol="features_raw")
stages.append(assembler)

# D. StandardScaler: Normalize features
# Ensures 'Live_Traffic_Index' (0-10) and 'Hour' (8-18) are on the same scale
scaler = StandardScaler(inputCol="features_raw", outputCol="features", withMean=True, withStd=True)
stages.append(scaler)

# 3. Apply the Pipeline
print("Running Transformation Pipeline...")
pipeline = Pipeline(stages=stages)
processed_df = pipeline.fit(df).transform(df)

# 4. Select Final Data for Modeling
# Spark ML expects 'features' (input) and 'label' (target)
final_data = processed_df.select(
    col("features"), 
    col("Actual_Delay_Minutes").alias("label")
)

# 5. Train-Test Split
print("Splitting Data into Train (80%) and Test (20%)...")
train_data, test_data = final_data.randomSplit([0.8, 0.2], seed=42)

print(f"✅ Training Set: {train_data.count()} rows")
print(f"✅ Testing Set:  {test_data.count()} rows")

# Preview
print("\n--- Processed Feature Vector (First 5 Rows) ---")
final_data.show(5, truncate=False)
```

```python
from pyspark.ml.regression import LinearRegression, RandomForestRegressor
from pyspark.ml.evaluation import RegressionEvaluator

print("=== STEP 3: MODEL TRAINING & EVALUATION (Fixed) ===\n")

# ==========================================
# 1. TRAIN LINEAR REGRESSION
# ==========================================
print("1. Training Linear Regression Model...")
lr = LinearRegression(featuresCol="features", labelCol="label", regParam=0.0)
lr_model = lr.fit(train_data)

# Predict
lr_predictions = lr_model.transform(test_data)

# ==========================================
# 2. TRAIN RANDOM FOREST
# ==========================================
print("2. Training Random Forest Model...")
# FIX: specific 'featuresCol="features"' (instead of features_raw)
rf = RandomForestRegressor(featuresCol="features", labelCol="label", numTrees=50, maxDepth=10, seed=42)
rf_model = rf.fit(train_data)

# Predict
rf_predictions = rf_model.transform(test_data)

# ==========================================
# 3. EVALUATION & COMPARISON
# ==========================================
print("\n=== MODEL COMPARISON RESULTS ===")

evaluator = RegressionEvaluator(labelCol="label", predictionCol="prediction")

# Linear Regression Metrics
lr_rmse = evaluator.evaluate(lr_predictions, {evaluator.metricName: "rmse"})
lr_r2 = evaluator.evaluate(lr_predictions, {evaluator.metricName: "r2"})

# Random Forest Metrics
rf_rmse = evaluator.evaluate(rf_predictions, {evaluator.metricName: "rmse"})
rf_r2 = evaluator.evaluate(rf_predictions, {evaluator.metricName: "r2"})

# Print Comparison Table
print("-" * 65)
print(f"{'Metric':<20} {'Linear Regression':<20} {'Random Forest':<20}")
print("-" * 65)
print(f"{'RMSE (Error)':<20} {lr_rmse:<20.4f} {rf_rmse:<20.4f}")
print(f"{'R² (Accuracy)':<20} {lr_r2:<20.4f} {rf_r2:<20.4f}")
print("-" * 65)

# ==========================================
# 4. CONCLUSION
# ==========================================
print("\n🏆 WINNER ANALYSIS:")
if lr_r2 >= rf_r2:
    print(f"✅ Linear Regression performed better (R²: {lr_r2:.2f}).")
    best_model = lr_model
    best_predictions = lr_predictions
else:
    print(f"✅ Random Forest performed better (R²: {rf_r2:.2f}).")
    best_model = rf_model
    best_predictions = rf_predictions

print("\n--- Sample Predictions (Best Model) ---")
best_predictions.select("label", "prediction").show(5)
```

```python
from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml.evaluation import RegressionEvaluator

print("=== STEP 3: RANDOM FOREST MODEL TRAINING ===\n")

# 1. Initialize the Random Forest Regressor
# - featuresCol: The column containing our feature vectors (from Step 2)
# - labelCol: The column containing the actual delay (target)
# - numTrees: Number of trees in the forest (50 is a good balance for speed/accuracy)
# - maxDepth: How deep each tree can go (10 captures complex patterns)
rf = RandomForestRegressor(featuresCol="features", labelCol="label", numTrees=50, maxDepth=10, seed=42)

# 2. Train the Model
print("Training the Random Forest model... (This might take a moment)")
rf_model = rf.fit(train_data)
print("✅ Model Trained Successfully!")

# 3. Make Predictions on Test Data
print("Generating predictions on the test set...")
predictions = rf_model.transform(test_data)

# 4. Evaluate the Model
evaluator = RegressionEvaluator(labelCol="label", predictionCol="prediction")

# Calculate RMSE (Root Mean Squared Error) - Lower is better
rmse = evaluator.evaluate(predictions, {evaluator.metricName: "rmse"})

# Calculate R² (R-Squared) - Higher is better (Target: > 0.60)
r2 = evaluator.evaluate(predictions, {evaluator.metricName: "r2"})

print("\n" + "="*40)
print(f"📊 MODEL PERFORMANCE METRICS")
print("="*40)
print(f"🚀 RMSE (Root Mean Squared Error): {rmse:.4f} minutes")
print(f"🚀 R² Score (Accuracy):          {r2:.4f}")
print("="*40)

# Check if we met the target
if r2 > 0.60:
    print("\n✅ SUCCESS: R² score is above 0.60!")
else:
    print("\n⚠️ WARNING: R² score is below 0.60. We might need to adjust the traffic simulation.")

# 5. Feature Importance (Optional but recommended)
# This tells us which features drove the predictions the most
print("\n🔍 Feature Importances (Top contributors to delay):")
print(rf_model.featureImportances)

# 6. Save the Model (Optional)
# model_path = "random_forest_bus_delay_model"
# rf_model.write().overwrite().save(model_path)
# print(f"\n💾 Model saved to: {model_path}")
```

```python
from pyspark.sql import Row
from pyspark.ml.feature import VectorAssembler

print("=" * 60)
print("🚌 REAL-TIME BUS DELAY PREDICTION SYSTEM 🚌")
print("=" * 60)

# 1. Get Unique Values for Dropdowns (Run once)
# This makes the menu dynamic based on your actual data
days_list = [row['Day_of_Week'] for row in df.select("Day_of_Week").distinct().collect()]
lines_list = [str(row['LineName']) for row in df.select("LineName").distinct().collect()]
# Sort them for better UX
days_list.sort(key=lambda x: ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"].index(x))
lines_list.sort()

continue_prediction = True

while continue_prediction:
    print("\n=== 📝 ENTER TRIP DETAILS ===\n")

    # --- INPUT 1: DAY OF WEEK ---
    print("Select Day:")
    for i, day in enumerate(days_list, 1):
        print(f"  {i}. {day}")
    try:
        day_choice = int(input("Enter number: ")) - 1
        selected_day = days_list[day_choice]
    except:
        print("❌ Invalid input. Defaulting to Monday.")
        selected_day = "Monday"

    # --- INPUT 2: BUS LINE ---
    print("\nSelect Bus Line:")
    for i, line in enumerate(lines_list, 1):
        print(f"  {i}. Line {line}")
    try:
        line_choice = int(input("Enter number: ")) - 1
        selected_line = lines_list[line_choice]
    except:
        selected_line = lines_list[0]
        print(f"❌ Invalid. Defaulting to {selected_line}")

    # --- INPUT 3: HOUR ---
    try:
        hour = int(input("\nEnter Hour (0-23): "))
        if hour < 0 or hour > 23: raise ValueError
    except:
        hour = 8
        print("❌ Invalid input. Defaulting to 8 AM.")

    # --- INPUT 4: LIVE TRAFFIC (Simulated Sensor) ---
    print("\nSimulate Live Traffic Condition:")
    print("   0 = Empty Roads")
    print("   5 = Moderate Traffic")
    print("  10 = Gridlock / Accident")
    try:
        traffic_input = float(input("Enter Traffic Index (0-10): "))
    except:
        traffic_input = 5.0
        print("❌ Invalid input. Defaulting to Moderate Traffic (5.0).")

    # ======================================================
    # DATA PREPARATION FOR MODEL
    # ======================================================
    
    # 1. Determine Route Type (Logic must match generation)
    # In our data, 597 is City Service, 598 is Commuter
    route_type = "City Service" if selected_line == '597' else "Commuter Express"

    # 2. Create a DataFrame for the User's Input
    user_df = spark.createDataFrame([
        Row(Hour_of_Day=hour, Day_of_Week=selected_day, 
            Route_Type=route_type, LineName=selected_line, 
            Live_Traffic_Index=traffic_input,
            Actual_Delay_Minutes=0) # Dummy value, needed for schema but ignored
    ])

    # 3. Transform the Input using the SAME Pipeline
    # We must fit the pipeline to the TRAINING data structure first to ensure indexers match
    # (In a real app, we would load a saved pipeline model)
    processed_input = pipeline.fit(df).transform(user_df)

    # 4. Make Prediction
    prediction = rf_model.transform(processed_input)
    predicted_delay = prediction.select("prediction").collect()[0][0]

    # ======================================================
    # DISPLAY RESULTS
    # ======================================================
    print("\n" + "=" * 60)
    print(f"🔮 PREDICTION RESULT For {selected_day} @ {hour}:00")
    print("=" * 60)
    print(f"🚦 Live Traffic Index: {traffic_input}/10")
    print(f"🚌 Bus Line: {selected_line} ({route_type})")
    print("-" * 60)
    print(f"⏱️  ESTIMATED DELAY: {predicted_delay:.1f} minutes")
    print("-" * 60)

    # Interpretation
    if predicted_delay < 5:
        print("✅ Status: ON TIME (Normal Operations)")
    elif predicted_delay < 15:
        print("⚠️ Status: MODERATE DELAY (Expect minor wait)")
    else:
        print("🚨 Status: SEVERE DELAY (Consider alternate route)")
    
    print("=" * 60 + "\n")

    # Loop
    check = input("Check another time? (y/n): ").lower()
    if check != 'y':
        continue_prediction = False
        print("\nExiting System. Safe Travels! 👋")
```
